```python
import pandas as pd
import numpy as np
import datetime
import time 
import seaborn as sns
from sklearn.model_selection import train_test_split
import sklearn.preprocessing
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1  = pd.read_excel('00907Supplemental_File_1_Harkey_et_al.xlsx',sheet_name = "ACC 449 Filtered Gene List")
data2 =  pd.read_excel('GoldStandard_ETR1Dependent_GenesLists.xlsx')
print('The shape of data1 is {}. The shape of data2 is {}.'.format(data1.shape, data2.shape))
```

```python
data1
```

```python
data1['ACC00_avg'] = data1[['ACC00_1','ACC00_2','ACC00_3']].mean(axis=1)
data1['ACC05_avg'] = data1[['ACC05_1','ACC05_2','ACC05_3']].mean(axis=1)
data1['ACC1_avg'] = data1[['ACC1_1','ACC1_2','ACC1_3']].mean(axis=1)
data1['ACC2_avg'] = data1[['ACC2_1','ACC2_2','ACC2_3']].mean(axis=1)
data1['ACC4_avg'] = data1[['ACC4_1','ACC4_2','ACC4_3']].mean(axis=1)
data1['ACC8_avg'] = data1[['ACC8_1','ACC8_2','ACC8_3']].mean(axis=1)
data1['ACC12_avg'] = data1[['ACC12_1','ACC12_2','ACC12_3']].mean(axis=1)
data1['ACC24_avg'] = data1[['ACC24_1','ACC24_2','ACC24_3']].mean(axis=1)
```

```python
df = data1[['ACC00_avg','ACC05_avg','ACC1_avg','ACC2_avg','ACC4_avg','ACC8_avg','ACC12_avg','ACC24_avg']]
df
```

```python
df = sklearn.preprocessing.minmax_scale(df, axis=1, copy=True)
```

```python
df = pd.DataFrame(df,columns = ['ACC00_avg','ACC05_avg','ACC1_avg','ACC2_avg','ACC4_avg','ACC8_avg','ACC12_avg','ACC24_avg'])
df
df.to_csv('averaged data.csv',index = False)
```
